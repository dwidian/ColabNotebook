## Common imports

```
import warnings
warnings.filterwarnings('ignore')
import numpy as np # numerical processing library
import matplotlib.pyplot as plt # chart library
%pylab inline
```

## Triangular membership function

```
def triangular(x, a, b, c):
    return max( min( (x-a)/(b-a), (c-x)/(c-b) ), 0 )
```

## Defition of the linguistic variable height

```
# U variable represents the Universe of Discourse of numerical variable height

U = np.arange(25,220,0.5) # generates a sequence of number between 0 and 10, step 0.2
```

```
height = {
  'Very Small': [25, 25.1, 50],
  'Small': [25, 50, 75],
  'Short': [50, 90, 130],
  'Medium': [90, 130, 175],
  'High': [130, 175, 220],
  'Very High': [175, 220, 220.1]
}
```

## Visualization of the fuzzy sets of the linguistic variable height

```
plt.figure(figsize=(15,7))

lines = []

for fuzzy_set in height.keys():
  memberships = [ triangular(x, *height[fuzzy_set]) for x in U ]
  # plot the chart
  tmp, = plot(U, memberships, label=fuzzy_set)
  lines.append(tmp)

legend(handles=lines)
xlabel("Height")
ylabel("Membership value")

    
```

## Fuzzyfication

```
for fuzzy_set in height.keys():
  print(fuzzy_set, triangular(163, *height[fuzzy_set]))
```
